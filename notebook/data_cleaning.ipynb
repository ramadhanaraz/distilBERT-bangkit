```python
  ## Dataset load
df = pd.read_csv("dataset.csv")
df.head(10)
```

```python
## Dropping column and check for NA values (tentuin kolom yang perlu dipake atau feature selection)
df_coldrop = df.drop(["geo", "job_board",  "post_date", "salary_offered","state","company_name", "url"], axis=1)
df_coldrop.head(10)
df_coldrop.query("category=='Trades & Services' and city=='Melbourne'")


##df_coldrop.isna()
##df_coldrop['salary_offered'] = df_coldrop['salary_offered'].replace(np.nan, ' ')
```

```python
## Cleaning data.. (kalo misalnya ada unicode character itu diilangin sama semua descriptionnya diubah jadi lower - case semua)

df_coldrop['category']=df_coldrop['category'].apply(str.lower)
df_coldrop['city']=df_coldrop['city'].apply(str.lower)
df_coldrop['job_description'] = df_coldrop['job_description'].astype(str)
df_coldrop['job_description']=df_coldrop['job_description'].apply(str.lower)
df_coldrop['job_title']=df_coldrop['job_title'].apply(str.lower)
df_coldrop['job_type']=df_coldrop['job_type'].apply(str.lower)
df_coldrop.head(10)

```
